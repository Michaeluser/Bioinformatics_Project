```python
def clinvar_score(clnsig):
    if pd.isna(clnsig):
        return 0
    
    s = clnsig.lower()
    
    if "pathogenic" in s:
        return 3
    elif "likely%20pathogenic" in s or "likely pathogenic" in s:
        return 2
    elif "uncertain" in s:
        return 1
    else:
        return 0

df2["clinvar_score"] = df2["ClinVar_SV_CLNSIG"].apply(clinvar_score)

df_sorted = df2.sort_values(
    ["Location", "SYMBOL", "clinvar_score"],
    ascending=[True, True, False]
)

df2 = df_sorted.drop_duplicates(subset=["Location", "SYMBOL"], keep="first")
```
